```python
def moving_average_crossover_strategy(data, initial_cash=100.0, buy_fee_rate=0.002, sell_fee_rate=0.002):
    """
    Hareketli ortalama crossover stratejisine g√∂re al-sat i≈ülemleri uygular.

    Parametreler:
        data: DataFrame (timestamp, close, ma_short, ma_long s√ºtunlarƒ± i√ßermeli)
        initial_cash: Ba≈ülangƒ±√ß USD bakiyesi
        buy_fee_rate: Alƒ±m i≈ülemi komisyon oranƒ± (√∂rn. 0.002 = %0.2)
        sell_fee_rate: Satƒ±≈ü i≈ülemi komisyon oranƒ± (√∂rn. 0.002 = %0.2)

    D√∂ner:
        trades_df: ƒ∞≈ülem kayƒ±tlarƒ±
        final_cash: Son USD deƒüeri
        total_return: Y√ºzdesel getiri
        total_fees: Toplam √∂denen i≈ülem √ºcretleri (USD cinsinden)
    """
    data = data.copy()

    data['signal'] = 0
    data.loc[data['ma_short'] > data['ma_long'], 'signal'] = 1
    data.loc[data['ma_short'] < data['ma_long'], 'signal'] = -1

    cash = initial_cash
    btc = 0.0
    total_fees = 0.0
    trades = []

    for i in range(1, len(data)):
        current = data.iloc[i]
        prev = data.iloc[i - 1]
        price = current['close']

        # Alƒ±m (BUY)
        if current['signal'] == 1 and prev['signal'] != 1 and cash > 0:
            fee = cash * buy_fee_rate
            effective_cash = cash - fee
            btc_amount = effective_cash / price
            btc += btc_amount
            total_fees += fee

            trades.append({
                'timestamp': current['timestamp'],
                'action': 'BUY',
                'price': price,
                'btc_amount': btc_amount,
                'cash': 0,
                'btc_balance': btc,
                'fee': fee
            })
            cash = 0

        # Satƒ±m (SELL)
        elif current['signal'] == -1 and prev['signal'] != -1 and btc > 0:
            gross_cash = btc * price
            fee = gross_cash * sell_fee_rate
            net_cash = gross_cash - fee
            total_fees += fee

            trades.append({
                'timestamp': current['timestamp'],
                'action': 'SELL',
                'price': price,
                'btc_amount': -btc,
                'cash': net_cash,
                'btc_balance': 0,
                'fee': fee
            })
            cash = net_cash
            btc = 0

    trades_df = pd.DataFrame(trades)
    final_value = cash + btc * data.iloc[-1]['close']
    total_return = (final_value - initial_cash) / initial_cash

    return trades_df, final_value, total_return, total_fees
```

```python
trades_df, final_cash, total_return, total_fees = moving_average_crossover_strategy(
    data=one_week_df,
    initial_cash=100.0,
    buy_fee_rate=0.001,
    sell_fee_rate=0.001
)

print(trades_df)
print(f"Final USD Bakiyesi: ${final_cash:.2f}")
print(f"Toplam Getiri: %{total_return * 100:.2f}")
print(f"Toplam √ñdenen Komisyon: ${total_fees:.4f}")
```

```python
def compute_rsi(data, period=14):
    """
    RSI (Relative Strength Index) hesaplar ve 'rsi' s√ºtununu data'ya ekler.
    """
    delta = data['close'].diff()
    gain = delta.clip(lower=0)
    loss = -delta.clip(upper=0)

    avg_gain = gain.rolling(window=period, min_periods=period).mean()
    avg_loss = loss.rolling(window=period, min_periods=period).mean()

    rs = avg_gain / avg_loss
    data['rsi'] = 100 - (100 / (1 + rs))

    return data
```

```python
def rsi_strategy(data, initial_cash=100.0, buy_fee_rate=0.002, sell_fee_rate=0.002, rsi_buy=30, rsi_sell=70):
    """
    RSI tabanlƒ± al-sat stratejisi uygular.

    RSI < rsi_buy ‚Üí Al
    RSI > rsi_sell ‚Üí Sat

    Parametreler:
        data: RSI i√ßeren DataFrame
        initial_cash: Ba≈ülangƒ±√ß nakit
        buy_fee_rate: Alƒ±m komisyon oranƒ±
        sell_fee_rate: Satƒ±≈ü komisyon oranƒ±

    D√∂nen:
        trades_df, final_cash, total_return, total_fees
    """
    data = data.copy()

    cash = initial_cash
    btc = 0.0
    total_fees = 0.0
    trades = []

    for i in range(1, len(data)):
        current = data.iloc[i]
        price = current['close']
        rsi = current['rsi']

        # RSI < 30 ‚Üí AL
        if rsi < rsi_buy and cash > 0:
            fee = cash * buy_fee_rate
            effective_cash = cash - fee
            btc_amount = effective_cash / price
            btc += btc_amount
            total_fees += fee

            trades.append({
                'timestamp': current['timestamp'],
                'action': 'BUY',
                'price': price,
                'btc_amount': btc_amount,
                'cash': 0,
                'btc_balance': btc,
                'fee': fee
            })
            cash = 0

        # RSI > 70 ‚Üí SAT
        elif rsi > rsi_sell and btc > 0:
            gross_cash = btc * price
            fee = gross_cash * sell_fee_rate
            net_cash = gross_cash - fee
            total_fees += fee

            trades.append({
                'timestamp': current['timestamp'],
                'action': 'SELL',
                'price': price,
                'btc_amount': -btc,
                'cash': net_cash,
                'btc_balance': 0,
                'fee': fee
            })
            cash = net_cash
            btc = 0

    trades_df = pd.DataFrame(trades)
    final_value = cash + btc * data.iloc[-1]['close']
    total_return = (final_value - initial_cash) / initial_cash

    return trades_df, final_value, total_return, total_fees
```

```python
def rsi_grid_search(data, initial_cash=100.0, buy_fee_rate=0.002, sell_fee_rate=0.002,
                    rsi_buy_range=range(20, 36, 5), rsi_sell_range=range(65, 86, 5), period=14):
    """
    RSI stratejisi i√ßin en iyi rsi_buy / rsi_sell kombinasyonlarƒ±nƒ± arar.

    D√∂nen:
        results_df: Kombinasyonlara g√∂re final deƒüer, getiri ve komisyon i√ßeren DataFrame.
    """

    # RSI hesapla
    data = compute_rsi(data.copy(), period=period)

    results = []

    for rsi_buy, rsi_sell in itertools.product(rsi_buy_range, rsi_sell_range):
        if rsi_buy >= rsi_sell:
            continue  # Ge√ßersiz kombinasyon

        trades_df, final_cash, total_return, total_fees = rsi_strategy(
            data=data,
            initial_cash=initial_cash,
            buy_fee_rate=buy_fee_rate,
            sell_fee_rate=sell_fee_rate,
            rsi_buy=rsi_buy,
            rsi_sell=rsi_sell
        )

        results.append({
            'rsi_buy': rsi_buy,
            'rsi_sell': rsi_sell,
            'final_cash': final_cash,
            'total_return_pct': total_return * 100,
            'total_fees': total_fees,
            'trade_count': len(trades_df)
        })

    results_df = pd.DataFrame(results)
    results_df = results_df.sort_values(by='final_cash', ascending=False).reset_index(drop=True)

    return results_df
```

```python
import seaborn as sns
import numpy as np
```

```python
def plot_rsi_grid_heatmap(results_df, value_col='final_cash'):
    """
    RSI grid sonu√ßlarƒ±nƒ± ƒ±sƒ± haritasƒ± (heatmap) olarak √ßizer.

    value_col: G√∂rselle≈ütirilecek metrik ('final_cash', 'total_return_pct', 'total_fees', vs.)
    """
    pivot = results_df.pivot(index='rsi_buy', columns='rsi_sell', values=value_col)

    plt.figure(figsize=(10, 6))
    sns.heatmap(pivot, annot=True, fmt=".2f", cmap="YlGnBu")
    plt.title(f"RSI Grid Search - {value_col.replace('_', ' ').title()}")
    plt.xlabel("RSI Sell Threshold")
    plt.ylabel("RSI Buy Threshold")
    plt.tight_layout()
    plt.show()
```

```python
# Grid search sonucunu alalƒ±m
results_df = rsi_grid_search(
    data=one_week_df,
    initial_cash=100.0,
    buy_fee_rate=0.001,
    sell_fee_rate=0.001,
    rsi_buy_range=range(20, 36, 5),
    rsi_sell_range=range(65, 86, 5),
    period=14
)

# Isƒ± haritasƒ±nƒ± √ßiz
plot_rsi_grid_heatmap(results_df, value_col='final_cash')  # ya da 'total_return_pct', 'trade_count', vs.
```
